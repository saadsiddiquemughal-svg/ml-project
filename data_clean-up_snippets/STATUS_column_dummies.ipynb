```python
import pandas as pd
```

```python
# Loading the dataset
df = pd.read_csv('../data/Train.csv')

# Create a copy of the data
df_copy = df.copy()
```

```python
# in column STATUS, change DEL, DEP and RTR to OTHER <--- here you decide which STATUS entries should be combined into entry OTHER
df_copy['STATUS'].replace({'DEL' :'OTHER', 'DEP' : 'OTHER', 'RTR' : 'OTHER'}, inplace=True)

# create dummies for STATUS column
status_dummies = pd.get_dummies(df_copy.STATUS)

# drop original STATUS column and newly created OTHER column
df_copy.drop(columns = 'STATUS', inplace = True)
status_dummies.drop(columns='OTHER', inplace= True)

# merge newly created status categorical dummies with original data frame (minus STATUS column)
df_copy = pd.concat([df_copy, status_dummies], axis=1)
```
