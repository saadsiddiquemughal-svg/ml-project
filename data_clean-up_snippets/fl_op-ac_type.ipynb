```python
train['fl_op'] = train['FLTID'].str[:3]

condition = [
    (train["fl_op"] == 'TU '),
    (train["fl_op"] == 'UG '),
    (train["fl_op"] == 'WKL'),
    (train["fl_op"] != 'TU ')
]

choices = [1, 2, 3, 4]

train["fl_op"] = np.select(condition, choices, default=np.nan)
train.columns
```
